# Step 1: Import scikit-learn

```python
import sklearn
```

 # Step 2: Import the dataset

```python
from sklearn.datasets import load_breast_cancer
```

## Loading the dataset

```python
data = load_breast_cancer()
```

## Creating new variables for each important set of information and assigning the data
> In other words, we can organize the data with the following commands:

```python
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']
```

## Printing class names

```python
print(label_names)
```

## Showing that label_names are mapped to binary values 0 and 1
> Here 0 represents malignant cancer and 1 represents benign cancer

```python
print(labels[0])
```

## Feature names

```python
print(feature_names[0])
```

## Feature values

```python
print(features[0])
```

# Step 3: Organizing data into sets

## Importing the train_test_split function from sklearn

```python
from sklearn.model_selection import train_test_split
```

## Splitting the data into training and test data
> Here we are using 40 % of the data for testing and the remaining data would be used for training
the model

```python
train, test, train_labels, test_labels = train_test_split(features,labels,test_size = 0.40, random_state = 42)
```

# Step 4: Building the model

## Importing the GaussianNB module of the NaÃ¯ve Bayes algorithm

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

## We will train the model by fitting it to the data by using gnb.fit()

```python
model = gnb.fit(train, train_labels)
```

# Step 5: Evaluating the model and its accuracy

## Making predictions on our test data

```python
preds = gnb.predict(test)
print(preds)
```

## Determining the accuracy of the model
> by comparing the two arrays namely test_labels and preds

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,preds))
```
